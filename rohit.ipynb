```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

%matplotlib inline
```

```python
df = pd.read_csv('employee_working_hours.csv')
```

```python
print("📌 First 5 Records:")
display(df.head())

print("\n📋 Data Info:")
df.info()

print("\n📈 Statistical Summary:")
display(df.describe())
```

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df['END_DATE'] = pd.to_datetime(df['END_DATE'])

df['DURATION_DAYS'] = (df['END_DATE'] - df['START_DATE']).dt.days + 1
df['AVG_HOURS_PER_DAY'] = df['NO_OF_HOURS'] / df['DURATION_DAYS']
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['NO_OF_HOURS'], bins=10, kde=True)
plt.title("Distribution of Total Working Hours")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='emp_name', y='NO_OF_HOURS', data=df, hue='department_name')
plt.title("Total Working Hours per Employee")
plt.xticks(rotation=45)
plt.ylabel("Total Hours")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='department_name', y='NO_OF_HOURS', data=df)
plt.title("Department-wise Distribution of Working Hours")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='emp_name', y='NO_OF_HOURS', data=df, hue='department_name')
plt.title("Total Working Hours per Employee")
plt.xticks(rotation=45)
plt.ylabel("Total Hours")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='department_name', y='NO_OF_HOURS', data=df)
plt.title("Department-wise Distribution of Working Hours")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['NO_OF_HOURS', 'DURATION_DAYS', 'AVG_HOURS_PER_DAY']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Numeric Features")
plt.show()
```

```python
print("\n🔍 Insights:")
print(f"Average Total Working Hours: {df['NO_OF_HOURS'].mean():.2f}")
print(f"Average Days Worked: {df['DURATION_DAYS'].mean():.2f}")
print(f"Average Hours Per Day: {df['AVG_HOURS_PER_DAY'].mean():.2f}")
```
